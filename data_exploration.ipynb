```python
import datetime
```

```python
api = openaq.OpenAQ()
today = datetime.datetime.today().date().strftime("%Y-%m-%d")
pm25 = api.measurements(country='US', date_from=today, parameter='pm25', df=True, limit=10000)
df = pm25.groupby('location').mean()
fig = px.scatter_geo(df, lat='coordinates.latitude', lon='coordinates.longitude', color='value', 
                        locationmode='USA-states', scope='usa', range_color=(0,50))
fig.show()
```
